## Bar Graphs
 - plt.bar()
 - plot a bar graph between heights and labels you provides. 

```python
x = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]

# x is an array of 4 numbers
```

```python
plt.bar(x, car1, width=0.3, tick_label= [2017, 2018, 2019, 2020], label="Car 1")
plt.bar(x+0.3, car2, width=0.3, label="Car 2")
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.xlim(-1, 5)
# plt.ylim()
plt.show()

# x: The positions on the x-axis where the bars will be placed.
# x+0.3 lgaya kyoki do bars car1 and car2 upar neeche aa rhe thee ek dusre ko ek saath plot krne pe. x + 0.3: Offsets the positions on the x-axis to avoid overlapping with the first set of bars. This creates a grouped bar plot.
# xlim stands for x limit. Iss se bars k ekdam left and ekdam right k space ka pta chalta kitna dena hai ie., padding jaisa hai ye.
```

## Pie Chart
- plt.pie()
- plot the pie chart with the provides labels and values

```python
subjects = ["Physics", 'Chemistry', 'Maths', "English"]
marks = np.array([78, 45, 90, 65])
```

```python
plt.pie(x= marks, labels = subjects, explode = [0, 0.1, 0, 0], shadow= True, autopct = "%0.0f")
plt.show()

# explode: This parameter specifies how far each slice is "exploded" or pulled out from the center of the pie. It is a list of the same length as marks. A value of 0 means the slice is not exploded, and a value of 0.1 means the slice is pulled out slightly. In this example, the second slice ("English") is pulled out slightly from the pie.
# 0.0f means after decimal 0 places. 0.2f means after decimal 2 digits.
```

## Histograms
 - plt.hist()
 - creating dummy data using normal distribution 

```python
X_snd = np.random.randn(500)
X_snd
# Random 500 numbers leliye. snd stands for standard normal distribution. Mean = 0, SD = 1
# np.random.randn: This function from the NumPy library generates samples from the standard normal distribution (a normal distribution with a mean of 0 and a standard deviation of 1).
```

```python
maths_mu = 70
maths_sig = 8
maths_marks = np.round(X_snd*maths_sig + maths_mu)

phy_mu = 50
phy_sig = 10
phy_marks = np.round( X_snd*phy_sig + phy_mu )

# SD(Standard Deviation) jo upar thaa usko increase krdiya maths_sig and phy_sig se multiply krke.
# Baad mein maths_mu and phy_mu ko add kiya taaki mean jo pahle 0 thaa vo increase ho jaaye.
```

```python
maths_marks
```

```python
plt.hist(maths_marks, bins=20)
plt.hist(phy_marks, bins = 10, alpha = 0.7)
plt.show()

# Dono histogram ek mein hi bana diye. Alpha means opacity.
# bins=10: Divides the data range into 10 intervals. bins=20: Divides the data range into 20 intervals. 10 bins mtlb 10 bars banege, 20 bins mtlb 20 bars banege uss given range mein.
```

```python
maths_marks.min()
```

```python
maths_marks.max()
```

```python
maths_marks.mean()
```

## Subplots
- create multiple plots in just a single figure.

```python
x = np.arange(0,10, 0.1)
y1 = x**2
y2 = x**3
y3 = np.sin(x)
y4 = np.cos(x)

# First Line -> 0 to 10 tak ka array bana diya with step size of 0.1
# 4 equations banadi 4 graphs banane k liye.
```

```python
plt.subplot(2,3,1)
plt.plot(x, y1)

plt.subplot(2,3,2)
plt.plot(x, y2)

plt.subplot(2,3, 3)
plt.plot(x, y3)

plt.subplot(2,3, 4)
plt.plot(x, y4)

plt.subplot(2,3, 5)
plt.plot(x, y3)

plt.subplot(2,3, 6)
plt.plot(x, y4)

plt.show()

# In subplot, 2 means 2 rows, 3 means 3 cols and last vala number shows index of graph i.e., konse number pe graph dikhana hai.
```

```python
a, b = np.meshgrid(a,b)

# The np.meshgrid function takes two 1D arrays, a and b, and returns two 2D matrices representing the coordinate grid defined by these arrays
# a mein jitni rows hai unko repeat krega b mein jitne cols hai utne time
# b mein jitne cols hai unko repeat krega a mein jitni rows hai utne time
```

```python
a
```

```python
fig = plt.figure()

# Creates a new figure object for plotting. This figure will hold the 3D plot.
```

```python
fig.add_subplot(projection ='3d')

# This adds a 3D subplot to the figure.
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.plot_surface(a, b, np.array([[5]]))
plt.show()

# 3rd line attempts to plot a surface using the x-coordinates from a, y-coordinates from b, and a z-coordinate of 5. z coordinate represents height.
```

```python
a = np.arange(-1, 1, 0.005)
b = a 
a, b = np.meshgrid(a, b)

# First line -> a is an array that starts at -1 and goes up to but does not include 1, with a step size of 0.005.
# Second Line -> This line simply assigns the array a to b. So, b will have the same values as a.
# Third Line -> np.meshgrid(a, b) will create two 2D arrays from the 1D arrays a and b.
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.plot_surface(a, b, a**2 + b**2)
plt.show()
```

## Visualize Iris Dataset
 - get uselful importance in a pictorial form.

```python
import pandas as pd
```

```python
iris = pd.read_csv('./iris.csv')
```

```python
iris.head()
```

```python
iris.shape
```

```python
petal_length = iris['petal_length']
petal_width = iris['petal_width']
```

```python
plt.scatter(x= petal_length, y = petal_width)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

```python
iris['species'] == 'setosa'

# Masking. Next step mein masking values ko unki real value mein badal diya.
```

```python
setosa = iris[ iris['species'] == 'setosa' ]
viriginca = iris[ iris['species'] == 'virginica' ]
versicolor = iris[ iris['species'] == 'versicolor' ]
```

```python
plt.scatter(setosa['petal_length'], setosa['petal_width'], color='green', label= "setosa")
plt.scatter(versicolor['petal_length'], versicolor['petal_width'], color='red', label="versicolor")
plt.scatter(viriginca['petal_length'], viriginca['petal_width'], color='yellow', label="virginica")
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.legend()
plt.show()

# petal_Length is for x-axis and petal_width is for y-axis.
```

```python
plt.hist(iris['petal_length'], bins = 20 )
plt.xlabel("petal length")
plt.ylabel("frequency")
plt.show()
```

```python
iris['species'].value_counts()
```

```python
plt.pie(iris['species'].value_counts(), labels= iris['species'].value_counts().index, autopct="%0.0f")
plt.show()
```
